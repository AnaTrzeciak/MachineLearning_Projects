Example of how to fit a linear dataset using linear regression with scikit-learn. Second Newton's Law a case of study.

Second Newton's Law is given by

$\vec{F} = m\times \vec{a}$

where $\vec{F}$ is the force vector, $m$ is the mass of the object and $\vec{a}$ is the acceleration.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
x_train = np.array([[54.79], [75.94], [93.02], [96.0], [139.53], [160.0]])
y_train = np.array([13140.84, 18093.17, 22849.37, 27703.64, 32705.01, 37657.34])
```

```python
print(f"x_train.shape: {x_train.shape}")
print(f"y_train.shape: {y_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Set the title
plt.title("Second Newton's Law")
# Set the y-axis label
plt.ylabel('F (dinas)')
# Set the x-axis label
plt.xlabel('a (cm/s²)')
plt.show()
```

```python
#Create and fit the regression model
#sgdr = SGDRegressor(max_iter = 300)
model = LinearRegression()
model.fit(x_train, y_train)
print(model)
#print(f"number of iterations completed: {model.n_iter_}, number of weight updates: {model.t_}")
```

```python
#Parameters updates
b = model.intercept_
w = model.coef_
print(f"model parameters: w: {w}, b: {b}")
```

```python
y_pred = model.predict(x_train)
print(y_pred)
R = model.score(x_train,y_train)
print(f"Score:{R}") #the best is score is 1.0
```

```python
fig,ax=plt.subplots(1,1,figsize=(9,5),sharey=True)
ax.scatter(x_train, y_train, marker='x', c='r', label = 'Real Data')
ax.plot(x_train, y_pred, c='blue', label ='Predict')
ax.set_ylabel('F (dinas)')
ax.set_xlabel('a (cm/s²)')
plt.legend()
plt.show()
```
